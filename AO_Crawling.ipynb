```python
trs = driver.find_element_by_xpath("/html/body/div[3]/div[2]/div[1]/div/div[3]/div/table/tbody").find_elements_by_tag_name("tr")


name_list = []
link_list = []
point_list = []


for tr in trs:
    see = tr.find_element_by_xpath('td[4]/a')
    
    name = see.text.strip()
    name_1 = name.split()
    first_name = name_1[0]
    first_name_s = first_name[0]
    last_name = list(name)
    del last_name[0:len(first_name)]
    last_name=''.join(last_name)
    last_name = last_name.strip()
    full_name = first_name_s + "."+last_name
    link = see.get_attribute('href')
    
    name_list.append(full_name)
    link_list.append(link)
    
    points = tr.find_element_by_xpath('td[6]/a').text
    points = int(points.replace(",",""))
    point_list.append(points)
```
